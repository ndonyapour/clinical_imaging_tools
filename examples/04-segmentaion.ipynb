```python
from vedo import Volume, show
import nibabel as nib

# Load the NIfTI file
img = nib.load("../data/nifti/dicom_0_11.nii.gz")
data = img.get_fdata()

# Visualize the volume
volume = Volume(data.transpose(1, 0, 2))
print(volume)
show(volume, axes=1, viewup="z", title="3D Volume Rendering")
```

```python
import matplotlib.pyplot as plt
import nibabel as nib 
from pathlib import Path

file_path = Path("../data/nifti/segmentations/skull.nii.gz") #segmentations/rib_left_9.nii.gz")
epi_img = nib.load(file_path)
epi_img_data = epi_img.get_fdata()

# Plot centeral slices 
# Get the shape of your image data
shape = epi_img_data.shape
print(f"Image shape: {shape}")

central_z = shape[2] // 2

plt.imshow(epi_img_data[central_z, :, :].T, cmap='gray')
plt.title(f'Central Sagittal Slice (z={central_z})')
plt.colorbar()
plt.axis('off')

plt.tight_layout()
plt.show()
```
